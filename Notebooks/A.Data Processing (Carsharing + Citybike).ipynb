## Carsharing-dataset

We found this interesting csv file on the open data portal austria (carsharing locations in Vienna). 

Datasource: https://www.data.gv.at/katalog/dataset/stadt-wien_universittenundfachhochschulenstandortewien/resource/c7f40490-9c28-4094-99ff-cc6afd492a82

Our idea was to create a dictionary with the carsharing-location ID´s (key) and the GPS coordinates (value) to reach our groupproject-goal.

### We want to rank the Universitaries in Vienna with the most closest Carsharing and Bikesharing (= Citybike) Locations.

```python
import os
import csv
```

```python
carsharing="./data/CARSHARINGOGD.csv"

import os
fSize = os.path.getsize(carsharing)

print('File size of'+carsharing+' is: '+str(fSize) + ' Bytes')
```

```python
with open(carsharing) as f:
    content=f.read()
    c=0
    for line in content.split("\n"): 
        c+=1
        if c<5:
            print("line"+str(c)+">> "+line)
print("total lines "+str(c))
```

```python
with open(carsharing) as f:
    csvfile = csv.reader(f)
    for i,row in enumerate(csvfile):
        if i<10:
            print(row[1],row[2])
```

We noticed that the GEO-coordinates are inverted (-y -x). 

```python
with open(carsharing) as f:
    csvfile = csv.reader(f)
    for i,row in enumerate(csvfile):
        if '4192' in row[1]:
            print(row)
```

```python
cargps = {}
with open(carsharing) as f:
    carsharingloc = csv.reader(f, delimiter=",")
    next(carsharingloc)
    for row in carsharingloc:
        y = row[2].strip().split("(")[1].split(" ")[0]
        x = row[2].strip().split("(")[1].split(" ")[1].split(")")[0]
        cargps[row[1]] = str(x + "," + y)
cargps
```

```python
len(cargps)
```

```python
caridadress = {}
with open(carsharing) as f:
    caridadressloc = csv.reader(f, delimiter=",")
    next(caridadressloc)
    for row in caridadressloc:
        caridadress[row[1] ]= row[4]
caridadress
```

```python
with open('./dictionaries/caridadress.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile, delimiter=',')
    for k, v in caridadress.items():
          writer.writerow( [ k, v] )
```
